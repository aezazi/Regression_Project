```python
%reset -fs
import bs4
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
import requests
import pickle
import re
```

```python
df = pd.read_pickle('df_all_movies_2016_pk')
df.head()
```

```python
row_count = df.shape[0]
row_count
```

```python
top_all_time_200_actors_df = pd.read_pickle('../top_all_time_200_actors_pk')
top_all_time_200_actors_df
```

```python

cast_2016_dict = {}

for i in range(row_count):
    cast_url = df.loc[i,'cast_page_url']
    
    r = requests.get(cast_url) #location in my dataframe
    s = bs(r.text)
    cast_rows = s.find(id = 'cast-and-crew').find_all('tr')
    
    cast_set = set()
    for row in cast_rows:
        person = row.find('a').text
        cast_set.add(person)
        
    cast_2016_dict[i] = cast_set
```

```python
cast_2016_dict[0]
```

```python
len(cast_2016_dict.keys())
```

```python


all_movies_2016_cast_list = []


for i in range(row_count):
    movie_cast = cast_2016_dict[i]
    cast_power_rating = 0
    
    for actor in movie_cast:
        if (actor in top_all_time_200_actors_df.values):
            print(actor)
            actor_rank = (top_all_time_200_actors_df.
            index[top_all_time_200_actors_df.Actor == actor].tolist()[0]) + 1
            print(actor_rank)
            actor_relative_rank_power = (201 - actor_rank)/200
            print(actor_relative_rank_power)
            cast_power_rating += actor_relative_rank_power
            cast_power_rating = round(cast_power_rating, 1)
            
            

    all_movies_2016_cast_list.append(cast_power_rating)
    print(cast_power_rating)
    print('-----------------------\n')

all_movies_2016_cast_list
```

```python
len(all_movies_2016_cast_list)
```

```python
cast_2016_df = pd.DataFrame(all_movies_2016_cast_list,
                                 columns = ['All_Cast_Rating'])
```

```python
cast_2016_df
```

```python
top10_actor_list = []
for i in range(10):
    actor = top_all_time_200_actors_df.iloc[i, 0]
    top10_actor_list.append(actor)
#     print(actor)
top10_actor_list
```

```python
for actor in top10_actor_list:
    cast_2016_df[actor] = 0
```

```python
cast_2016_df
```

```python
for i in range(row_count):
    movie_cast = cast_2016_dict[i]
    
    for actor in movie_cast:
        if (actor in top10_actor_list):
            cast_2016_df.loc[i, actor] = 1
    
```
