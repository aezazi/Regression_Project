```python
#Extract top 100 actors and add to top all_time_200 list

r = requests.get('https://www.the-numbers.com/box-office-star-records/domestic/lifetime-acting/top-grossing-leading-stars')
s = bs(r.text)

top_all_time_200_actors_list = []

ex_chart = s.find_all(id = 'page_filling_chart')[1].find_all('tr')

for i in range(1,101):
    actor = (ex_chart[i].find_all('a')[0].text)
#     print(actor)
    top_all_time_200_actors_list.append(actor)
    
len(top_all_time_200_actors_list)
# top_all_time_200_actors_set



#Extract actors 101 to all_time_200 and add to top all_time_200 list

r = requests.get('https://www.the-numbers.com/box-office-star-records/domestic/lifetime-acting/top-grossing-leading-stars/101')
s = bs(r.text)

ex_chart = s.find_all(id = 'page_filling_chart')[1].find_all('tr')

for i in range(1,101):
    actor = (ex_chart[i].find_all('a')[0].text)
#     print(actor)
    top_all_time_200_actors_list.append(actor)
    


len(top_all_time_200_actors_list)
```

```python
top_all_time_200_actors_df = pd.DataFrame(top_all_time_200_actors_list, columns = ['Actor'])
top_all_time_200_actors_df
top_all_time_200_actors_df.to_pickle('top_all_time_200_actors_pk')
```
